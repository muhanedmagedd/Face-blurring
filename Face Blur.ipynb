```python
import cv2
import os
import argparse
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe import Image
from mediapipe.tasks.python import vision, BaseOptions
```

```python
options = vision.FaceDetectorOptions(
        base_options= python.BaseOptions(model_asset_path='blaze_face_short_range.tflite'),
        running_mode= vision.RunningMode.IMAGE,
        min_detection_confidence=0.5
)
with vision.FaceDetector.create_from_options(options) as detector:
    image_path = r'D:\Computer Vision\Face Blurring\360_F_610194339_3CtGOkv4wIiAyybcib4IrFX0nnc83Bv6.jpg'
    img = cv2.imread(image_path)
    RGB_Img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
    mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=RGB_Img)
    result_detection = detector.detect(mp_image)

    if result_detection.detections is not None:
        for detection in result_detection.detections:
            bbox = detection.bounding_box
            X = int(bbox.origin_x)
            Y = int(bbox.origin_y)
            Width = int(bbox.width)
            Height = int(bbox.height)

            img[Y:Y+Height , X:X+Width , :] = cv2.blur(img[Y:Y+Height , X:X+Width , :],(75,75))
            cv2.imwrite(os.path.join('D:\Computer Vision\Face Blurring','imout.jpg'),img)

```

```python
vid_path = 'D:\Computer Vision\Face Blurring\WhatsApp Video 2026-03-30 at 5.38.38 PM.mp4'
output_dir = 'D:\Computer Vision\Face Blurring'
options1 = vision.FaceDetectorOptions(
    base_options = python.BaseOptions(model_asset_path='blaze_face_short_range.tflite') ,
    running_mode = vision.RunningMode.VIDEO,
    min_detection_confidence = 0.5
)
with vision.FaceDetector.create_from_options(options1) as detector1:
    video = cv2.VideoCapture(vid_path)
    ret,frame = video.read()
    Output_vid = cv2.VideoWriter(os.path.join(output_dir,'outputvid.mp4'),cv2.VideoWriter.fourcc(*'MP4V'),25,(frame.shape[1],frame.shape[0]))

    frame_time = int(1000/video.get(cv2.CAP_PROP_FPS))
    timestamp_ms = 0
    while ret:
        RGB_frame = cv2.cvtColor(frame , cv2.COLOR_BGR2RGB)
        mp_format = mp.Image(image_format=mp.ImageFormat.SRGB , data=RGB_frame)
        result = detector1.detect_for_video(mp_format,timestamp_ms)

        timestamp_ms += frame_time
        if result.detections is not None:
            for detection in result.detections:
                bbox1 = detection.bounding_box
                X1 = int(bbox1.origin_x)
                Y1 = int(bbox1.origin_y)
                H1 = int(bbox1.height)
                W1 = int(bbox1.width)
                frame[Y1:Y1+H1 , X1:X1+W1 , :] = cv2.blur(frame[Y1:Y1+H1 , X1:X1+W1 , :] , (50,50))
            # لو بيعمل blur ل 3 وشوش كل ما هيعمل بلور هيعمل save لنفس الفريم 3 مرات لو write كان جوه ال for loop
            Output_vid.write(frame)
        ret, frame = video.read()
    video.release()
    Output_vid.release()

```

```python
options2 = vision.FaceDetectorOptions(
    base_options = python.BaseOptions(model_asset_path='blaze_face_short_range.tflite'),
    running_mode = vision.RunningMode.IMAGE ,
    min_detection_confidence = 0.5,
)

with vision.FaceDetector.create_from_options(options2) as detector2:
    webcam = cv2.VideoCapture(0)
    while True:
        _,frame = webcam.read()
        RGB_frame = cv2.cvtColor(frame , cv2.COLOR_BGR2RGB)
        MP_frame = mp.Image(image_format = mp.ImageFormat.SRGB , data=RGB_frame)
        # لازما اللي داخل ع detector او mp عموما يكون بصيغته mp format لكن في العرض باخد frame عادي bgr
        # لكن لو بعمل write بال writer بيكون يفضل BGR
        result = detector2.detect(MP_frame)
        for detection in result.detections:
            bbox2 = detection.bounding_box
            X2,Y2,H2,W2 = int(bbox2.origin_x), int(bbox2.origin_y), int(bbox2.height),int(bbox2.width)
            frame[Y2:Y2+H2 ,X2:X2+W2 , :] = cv2.blur(frame[Y2:Y2+H2 ,X2:X2+W2 , :] , (75,75))
        cv2.imshow("Live",frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    webcam.release()
    cv2.destroyAllWindows()
```
